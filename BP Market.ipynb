# MARKET ANALYSIS SPORTSWEAR AND ATHLEISURE MARKET

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

plt.rcParams["font.family"] = "Times New Roman"
```

```python
path = "Market Data/Sportwear Market Size Europe.xlsx"
market = pd.read_excel(path)
```

```python
market = market[["Geography","Year", "Revenues", "y-o-y ex rates"]]
market_W = market[market["Geography"] == "World"]
market_W
```

```python
market_WEU = market[market["Geography"] == "Western Europe"]
market_WEU 
```

```python
market_EEU = market[market["Geography"] == "Eastern Europe"]
market_EEU
```

```python
market_EU = market[market["Geography"] == "Europe"]
market_EU
```

## Sportswear: Revenues Evolution and YOY by Geography of Interest


```python
# Plotting Revenue Evolution 
fig,ax = plt.subplots(figsize=(15,5))

colors = ["#010B40","#D5C8CE","#000000","#D94929"]
colors =  ["#AAA7F2", "#F2F2F2", "#CAF2C9", "#A6A08F"]
# Plot 1: Year/month vs. Total Orders
ax = sns.lineplot(x="Year", y="Revenues", hue="Geography", palette = colors, marker="o", data = market, lw=1)
# Set x-axis label
ax.set_xlabel("Year", fontsize=15, labelpad = 20,fontname = "times")
plt.xticks(rotation = 45, fontsize=15, fontname = "times")
# Set y-axis label
ax.set_ylabel("Retail Value - EUR million", fontsize=15, labelpad = 30, fontname = "times")
plt.yticks(fontsize=20, fontname = "times")
# Remove gridlines
plt.grid(False)
plt.title('Retail Value - Comparison among Geographies', fontsize = 15, fontname = "times")
plt.legend(fontsize=15)
ax.axvline(x=2020, linewidth=1, color='black')
plt.savefig("Graphs/retail value_comparison by geography.png", bbox_inches='tight', dpi=300)
plt.show()
```

```python
market[market["Geography"] == "Europe"]
```

```python
# Plot YOY by Year

plt.figure(figsize=(17,5))

sns.set(style="whitegrid")
colors =  ["#AAA7F2", "#F2F2F2", "#CAF2C9", "#A6A08F"]
ax = sns.lineplot(x='Year', y="y-o-y ex rates",hue="Geography", data = market, 
                  palette = colors, marker="o")
ax.set_xlabel("Year",fontsize=12, labelpad = 20)
ax.set_ylabel("% Y-O-Y Growth",fontsize=12, labelpad = 20)
plt.xticks(rotation = 90, fontsize = 12)
plt.yticks(fontsize = 12)
plt.title('YoY rate - Comparison among Geographies', fontsize = 15)

ax.axhline(linewidth=1, color='black')
ax.axvline(x=2020, linewidth=1, color='black')
plt.grid(False)
plt.savefig("Graphs/YoY_comparison by geography.png", bbox_inches='tight', dpi=300)
plt.show()
```

## Sportswear:Revenues Evolution and YOY by Country


```python
path = "Market Data/Sportwear Market Size Europe.xlsx"
market_bycountry = pd.read_excel(path,"European Countries")
market_bycountry_EEU = market_bycountry[market_bycountry["Region"] == "Eastern Europe"]
market_bycountry_WEU = market_bycountry[market_bycountry["Region"] == "Western Europe"]
```

```python
market_bycountry_EEU_2020_2024 = market_bycountry_EEU[(market_bycountry_EEU["Year"] == 2020) | (market_bycountry_EEU["Year"] == 2024)]

plt.figure(figsize=(20,8))
colors =  ["#AAA7F2", "#F2F2F2", "#CAF2C9"]
sns.set_palette(colors)

sns.barplot(x="Geography", hue="Year", y="Retail Value - EUR Million", data=market_bycountry_EEU_2020_2024)
plt.show()
```

```python
plt.figure(figsize=(16,7))

sns.set(style="whitegrid")
colors = ["#AAA7F2", "#F2F2F2", "#CAF2C9", "#A6A08F", "#D9C58B","#D9D5D2"]
ax = sns.lineplot(x='Year', y="YoY",hue="Geography",data = market_bycountry_EEU, 
                  palette = colors, marker="o")
ax.set_xlabel("Year",fontsize=12, labelpad = 20)
ax.set_ylabel("% Y-O-Y Growth",fontsize=12, labelpad = 20)
plt.xticks(rotation = 90, fontsize = 12)
plt.yticks(fontsize = 12)
plt.title('YoY rate - Easter Europe by Country', fontsize = 15)
plt.savefig("Graphs/YoY - Easter Europe by Country.png", bbox_inches='tight', dpi=300)

ax.axhline(linewidth=1, color='black')
ax.axvline(x=2020, linewidth=1, color='black')
plt.grid(False)
plt.show()
```

```python
forecast_EEU_avggrowth = market_bycountry_EEU[(market_bycountry_EEU["Year"] == 2021) |(market_bycountry_EEU["Year"] == 2022) | (market_bycountry_EEU["Year"] == 2023) |(market_bycountry_EEU["Year"] == 2024)]
forecast_EEU_avggrowth = forecast_EEU_avggrowth.groupby(["Geography"]).mean().reset_index()
forecast_EEU_avggrowth = forecast_EEU_avggrowth[["Geography", "YoY", "Retail Value - EUR Million"]]
forecast_EEU_avggrowth = forecast_EEU_avggrowth.sort_values(by=["Retail Value - EUR Million"], ascending = False)
forecast_EEU_avggrowth.reset_index().drop(columns = "index")
```

## Sportswear: Analysis by Category

```python
path = "Market Data/Sportwear Market Size Europe.xlsx"
WE_bycat = pd.read_excel(path,"Western Europe - SW by Category")
WE_bycat
```

```python
WE_bycat["Expected Retail Value 2024 - EUR Million"] = WE_bycat.apply(lambda row: row["Category Value"] * ((1 + (row["%CAGR {Forecast}"] / 100)) **4), axis = 1)
WE_bycat
```

```python
# Plotting Revenue Evolution 
fig = plt.figure(figsize=(20,10))

plt.rcParams['axes.labelsize'] = 16
plt.rcParams['axes.titlesize'] = 16

csfont = {'fontname':'Times New Roman'}
colors = ["#010B40","#D5C8CE","#BF8C88","#D94929", "#F2F2F2","#323A40","#C7BAD9", "#D9D9D9"]
colors = ["#F2F2F2","#F2F2F2","#F2F2F2","#AAA7F2","#F2F2F2","#F2F2F2","#F2F2F2", "#F2F2F2"]

plt.bar(WE_bycat["Category"], WE_bycat["Category Value"], color = colors, width=0.8, bottom=None, align='center', data=WE_bycat)
plt.ylabel('Retail Value by Category - EUR million', **csfont, fontsize = 20)
plt.xticks(rotation = 20, fontsize = 20, fontname = "times")
plt.yticks(fontsize = 20, fontname = "times")

x = WE_bycat["Category"]
y = WE_bycat["%CAGR {Forecast}"]
axes2 = plt.twinx()
axes2.plot(x, y, color='k')
axes2.set_ylabel('%CAGR {Forecast}', **csfont, fontsize = 20)
plt.savefig("Graphs/Retail & YoY Spain by cat - WE.png", bbox_inches='tight', dpi=300)
plt.ylim(1, 8);

plt.show()
```
